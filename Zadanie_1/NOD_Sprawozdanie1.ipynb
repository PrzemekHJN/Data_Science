```python
# Zadanie - Łączenie ramek za pomocą metod concat() i merge()
# creating example dataframes
df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Name": ["Przemek", "Andrzej", "Kamil", "Johnny", "Piotrek"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 4, 8, 9],
    "University": ["UBB", "UŚ", "UJ", "UEK", "UP"]
})

# merging on ID column
df_merge = pd.merge(df1, df2, on="ID", how="inner")
print(df_merge)
```

```python
# Zadanie - Dodawanie nowych kolumn za pomocą operacji matematycznych
# add new column with sumed age of all childern for a parent
df_csv["total_child_age_months"] = (
    df_csv["u2_child1_age_months"].fillna(0) +
    df_csv["u2_child2_age_months"].fillna(0) +
    df_csv["u2_child3_age_months"].fillna(0)
)
```

```python
# Zadanie - Przykład dodawania nowych kolumn za pomocą funkcji lambda
# using lampda function to add new column - age category
df_csv["age_group"] = df_csv["age"].apply(
    lambda x: "very young parent" if "16 to 25" in str(x)
    else ("young parent" if "26 to 35" in str(x) else "inna grupa")
)
df_csv["age_group"]
```

```python
# Zadanie - Możliwości pracy z dużymi plikami przy użyciu argumentu chunksize
# using chunksize argument
chunk_stats = {}

for chunk in pd.read_csv("IHME_PREM_CH_HEALTH_2020_Y2021M10D11.csv", chunksize=5000):
    temp = chunk.groupby("country")["u2_pre_no_provider_why"].apply(
        lambda x: (x == "Health facility closed").sum()
    )
    for country, count in temp.items():
        chunk_stats[country] = chunk_stats.get(country, 0) + count

print(chunk_stats)
```
